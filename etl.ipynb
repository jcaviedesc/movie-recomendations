```python
import pandas as pd
import ast
```

## carga de datos 

```python
df = pd.read_csv('Dataset/movies_dataset.csv')

# Imprimir el nombre de las columnas
print(df.info)

# describo los datos
print(df.describe())

```

```python
def safe_literal_eval(x):
    try:
        return ast.literal_eval(x) if pd.notnull(x) else {}
    except Exception:
        return {} # si no podemos convertir a diccionario retornamos uno vacio
    
def get_dic_value_by_key(key):
    def get_value(x):
        try:
            return x.get(key, '')
        except Exception:
            return ''
    return get_value
```

```python
# Convertir la columna 'belongs_to_collection' de cadena a diccionario
df['belongs_to_collection'] = df['belongs_to_collection'].fillna('{}')
# print(df.iloc[0])
df['belongs_to_collection'] = df['belongs_to_collection'].apply(safe_literal_eval)
# Crear nuevas columnas a partir de la clave del diccionario
# df1 = pd.json_normalize(df['belongs_to_collection'].map(eval)) 

df['belongs_to_collection_id'] = df['belongs_to_collection'].apply(get_dic_value_by_key('id'))
df['belongs_to_collection_name'] = df['belongs_to_collection'].apply(get_dic_value_by_key('name'))
# print(df[['belongs_to_collection']].info())

print(df[['belongs_to_collection_id', 'belongs_to_collection_name']])

```

```python
def get_dic_value_from_list(key):
    def map_key(x):
        try:
            return', '.join([str(genre.get(key,'') ) for genre in x])
        except Exception:
            print(f'key {key} with val {x} is type {type(x)}')
            return '' 
        
    return map_key
```

```python
# Convertir la columna 'genres' de cadena a una lista de diccionarios
df['genres'] = df['genres'].apply(lambda x: eval(x) if pd.notnull(x) else [])
# Crear una nueva columna 'genres_name' con la concatenación de las claves 'name' de los objetos
df['genres_name'] = df['genres'].apply(get_dic_value_from_list('name'))
# Crear una nueva columna 'genres_id' con la concatenación de las claves 'name' de los objetos
df['genres_id'] = df['genres'].apply(get_dic_value_from_list('id'))

df[['genres_name', 'genres_id']]
```

```python
print(df['production_companies'])
# Convertir la columna 'production_companies' de cadena a una lista de diccionarios
df['production_companies'] = df['production_companies'].apply(lambda x: eval(x) if pd.notnull(x) else [])
# Crear una nueva columna 'production_companies_name' con la concatenación de las claves 'name' de los objetos
df['production_companies_name'] = df['production_companies'].apply(get_dic_value_from_list('name'))
# Crear una nueva columna 'production_companies_id' con la concatenación de las claves 'name' de los objetos
df['production_companies_id'] = df['production_companies'].apply(get_dic_value_from_list('id'))

df[['production_companies_name', 'production_companies_id']]
```

```python
#spoken_languages
print(df['spoken_languages'])
# Convertir la columna 'spoken_languages' de cadena a una lista de diccionarios
df['spoken_languages'] = df['spoken_languages'].apply(lambda x: eval(x) if pd.notnull(x) else [])
# Crear una nueva columna 'spoken_languages_name' con la concatenación de las claves 'name' de los objetos
df['spoken_languages_name'] = df['spoken_languages'].apply(get_dic_value_from_list('name'))
# Crear una nueva columna 'spoken_languages_id' con la concatenación de las claves 'name' de los objetos
df['spoken_languages_code'] = df['spoken_languages'].apply(get_dic_value_from_list('iso_639_1'))

df[['spoken_languages_name', 'spoken_languages_code']]
```

```python
#Los valores nulos de los campos revenue, budget deben ser rellenados por el número 0
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce').fillna(0).astype("Float32")
print(df["revenue"])
df["budget"] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype("Float32")
print(df["budget"])
```

```python
#Los valores nulos del campo release_date deben eliminarse.
num_datos = len(df)
print("numero de datos", len(df))
df=df.dropna(subset=['release_date'])
print("numero de datos borrados despues de borrar filas con valores nulos de release_date", num_datos- len(df))
df['release_date']
```

```python
# De haber fechas, deberán tener el formato AAAA-mm-dd, además deberán crear la columna release_year 
# donde extraerán el año de la fecha de estreno.

# Convierto la columna 'release_date' al tipo de dato datetime para poder obtener el año
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors="coerce")

# borro los datos que no pueden ser convertidos a fecha
df=df.dropna(subset=['release_date'])

# creo la columna 'year' con el año extraído de 'release_date'
df['year'] = df['release_date'].dt.year

df['year']
```

```python
'''
Crear la columna con el retorno de inversión, llamada return con los campos revenue y budget,
dividiendo estas dos últimas revenue / budget, cuando no hay datos disponibles para calcularlo, 
deberá tomar el valor 0.
'''
#creo la columna retorno de inversión 
df['return_of_investment'] = df.apply(lambda row: row['revenue'] / row['budget'] if row['revenue'] > 0 and row['budget'] > 0 else 0, axis=1)
```

```python
# Eliminar las columnas no utilizadas
columnas_no_utilizadas = ['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage']
df = df.drop(columnas_no_utilizadas, axis=1)
df
```

```python
# guardo la dataset transformado en un file .csv
df.to_csv('movies_dataset_processed.csv', index=False)
```
